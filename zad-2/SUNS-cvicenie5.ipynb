# Cvičenie č. 5

Zadanie dnešného cvičenia je stručnejšie, budete pracovať s rovnakou dátovou sadou, s rovnakými metódamy vyhodnotenia (zatiaľ), zmenia sa iba modely, ktoré trénujete.

1. Načítajte dáta, kódery a normalizéry prichistané na cvičení číslo 2.

2. Pre obe úlohy z predošlého cvičenia (klasifikácia aj regresia) natrénujte nové modely (SVM, strom, les, ...). Porovnajte dosiahnuté výsledky modelov. Zatiaľ môžete pracovať s defaultnými nastaveniami jednotlivých modelov, vyskúšajte si ich viacero (minimálne tie, ktoré sú v zozname, ale na sklearn nájdete aj ďalšie).
